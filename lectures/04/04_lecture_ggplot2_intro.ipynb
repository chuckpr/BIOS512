1. Start with the `ggplot` function
1. Encodings are defined with the `aes` instuction/function
1. We pass in the data as the first "argument"
1. We pass sncodings to `ggplot` in the second argument
1. Capture the output


````r
encodings = aes(x = displ, y = hwy)

p = ggplot(mpg, encodings) # mpg is data, where do "hwy" and "displ" come from

p
````

Add the `geom` you want to represent the data.

````r
p = p + geom_point(size = 2)

p
````

What if we want to color the points by `class`?

````r
encodings = aes(x = displ, y = hwy)

p = ggplot(mpg, encodings)

p = p + geom_point(size = 2)

p
````

````r
# shorthand...

p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point(size = 2)

p
````

````r
# more shorthand...

p = ggplot(mpg, aes(x = displ, y = hwy)) + 
    geom_point(size = 2)

p
````
