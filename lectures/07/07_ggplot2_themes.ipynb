## `labs`

- `axis.title`

- if we're changing text we use `element_text`

```r
# ?element_text
```

```r
options(repr.plot.width = 4, repr.plot.height = 3)

p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme(axis.title.x = element_text(color = 'red'))

p
```

## axis "text" (some times referred to as "ticks labels")

```r
p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme(axis.text = element_text(color = "orange"))

p
```

## "plot" and the "panel"

- for plot and panel we use `element_rect`

```r
p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme(panel.background = element_rect(fill = 'lightblue')) # try with panel

p
```

## "grid" and "ticks"

- for the grid and the ticks we use `element_line`

```r
p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme(
    axis.ticks.x = element_line(color = 'blue'),
    axis.title = element_blank(),
    panel.grid.major = element_line(color = "gray90"),
    panel.background = element_blank()
) # try with axis.ticks

p
```

- use `element_blank` to turn something off

## Margins

- to adjust margins use `margin`
- the order is top, right, bottom, left (i.e. clockwise starting from the top)

```r
p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme(
    axis.title.x = element_text(margin = margin(t = 12)),
    axis.title.y = element_text(margin = margin(r = 12))
) # try with panel

p
```

```r
p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point()

p = p + labs(x = "Displacement, mL", y = "Highway, MPG")

p = p + theme_bw()

p = p + theme(
    axis.title.x = element_text(margin = margin(t = 12)),
    axis.title.y = element_text(margin = margin(r = 12))
)

p
```
